Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans_1 : Matplotlib is a python library used to create visuals of the given data points.There are multiple
plot that can be plotted using the Pyplot module of matplotlib. That are:

    1. bar graph
    2. scatter plot
    3. pie charts
    4. line plot
    5. histograms
    

Ans_2 : Scatter plot are the graph that represents the relationship between two variables in a data set. t. It represents data points on a two-dimensional plane or on a Cartesian system.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))

plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Title")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.


Ans_3 : The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure.
It allows you to arrange and display multiple plots in a grid-like fashion, making it easier to visualize and
compare different datasets or aspects of your data.

Ans_4 : A bar plot, also known as a bar chart or bar graph, is a data visualization technique used to represent 
categorical data with rectangular bars. Each bar in a bar plot typically represents a category, and the height 
or length of the bar is proportional to the value or frequency of that category.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

Ans_5 : A box plot, also known as a box-and-whisker plot, is a data visualization technique used to display the
summary statistics and distribution of a dataset. It provides a graphical representation of the central tendency,
spread, and potential outliers within the data.

```python
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
```
