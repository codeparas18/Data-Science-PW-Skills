Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans_4 : LookupError raised whenever an invalid loopup operation is performed, mostly while accessing keys that
are non-existent in a dictionary or a set.
Example : 


```python
# Index Error
lst = ['mango','pineapple','banana','cheeku']

try:
    print(lst[5])
except Exception as e:
    print("Index Error : ",e)
else:
    print("Program successfully executed")
```

```python
# Key Error
lst = {'mango' : 1 ,'pineapple' : 2, 'cheeku' : 3}

try:
    value  = lst['banana']
    print(value)
    
except KeyError as e:
    print("KeyError : Key does not found")
    
else:
    print("Program successfully executed")
```

Q6. List down some best practices for exception handling in python.

Ans_6 : Some best practices for exception handling in Python are : 
    1) Never use exceptions for flow-control
    2) Handle exceptions at the level that knows how to handle them.
    3) Document or log the exceptions thrown by your code
